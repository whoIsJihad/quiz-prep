```python
import numpy as np
import matplotlib.pyplot as plt 
np.zeros((4,4,5)) 
#creates a 3d numpy array initialised as zero 
#total 4 2D arrays with 4 rows and 5 cols 

np.zeros(3) 
# creates a 1d array of size 3 initialised as zero


np.ones(3)
# creates a 1d array of size 3 with initialised as one 


np.eye(3)
#creates a identity matrix of size 3 x 3


np.arange(0,5,1)
#creates an array of numbers starting from 0 and ending at 5 (exclusive) with step size of 1


np.linspace(0,3,100)

#creates an array of numbers starting from 0 and ending at 3 (inclusive).length of the array 100


np.random.rand(3)

#random values in [0,1]

np.random.rand(3,4)
#random values in [0,1]but this time it is a 2d array 


a=3*np.random.rand(5)
#random values in [0,3]

np.random.randint(1,10,(2,3))
#creates random values from 1 to 10 and stores them in a 2d array where the dim of the array is 2,3


```

```python
arr=np.array([[2,3,4],[4,5,6],[6,7,9]])
print(arr)
#transpose of a matrix
print(arr.T)

#matrix multiplication
print(np.matmul(arr,arr.T))

#determinant 
print(np.linalg.det(arr))
print(np.linalg.det(arr.T))
```

```python

#array properties 
arr = np.array([[1, 2, 3], [3, 4, 5]])

print("Shape of the array:", arr.shape)
# shape of the array (2, 3)

print("Dimensions of the array:", arr.ndim)
# dimensions of the array = 2

print("Number of elements in the array:", arr.size)
# number of elements in the array = 6

print("Data type of the array:", arr.dtype)
# data type of the array


print("Total bytes consumed by the array :",arr.nbytes)

```

```python
arr=np.array([[2,3,4],[4,5,6],[6,7,9],[5,6,7]])
#create a 2d array of shape 4 x 3

arr2=arr.copy()
#make a copy of it

#reshaping the array in a 3 x 4 array
print(arr2.reshape(3,4))

print(arr.flatten())
print(arr.squeeze())
```

```python
#indexing and slicing 

arr=100*np.random.rand(4,5)
print(arr)


#access all rows of col 0
print(arr[:,0])

#access all colums of row 2

print(arr[2,:])

#slice subarray 

sliced_arrray=arr[1:3,2:4]
print(sliced_arrray)
```

```python
#element wise operation
arr=np.array([34,53,223],dtype=np.float64)
print(arr)
#both the operation does the same thing 
print(arr+5)
print(np.add(arr,5))


#multiply array by a constant 
print(arr*2)
print(np.multiply(2,arr))
print(np.multiply(arr,2))

#power of an element
arr2=arr**10
arr2
```

```python
#common built in fucntions 
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
#standard deviation of the array 

print(np.std(arr))
print(np.var(arr))
#min and max of the array
print(np.min(arr))
print(np.max(arr))

#index of max and min value
long_array=np.array([2,3,4,4,5,6,34,1,10,4,5,6])
print(np.argmax(long_array))
print(np.argmin(long_array))

#aggregation along axis 
arr=np.array([[2,3,4],[4,5,6],[6,7,9],[5,6,7]])

#sum along colums
print(np.sum(arr,axis=0))

#sum acroos rows 
print(np.sum(arr,axis=1))
```

```python
#np.sort(array) returns a new arraythat is sorted 
a=np.array([2,4,8,16])
a=np.sort(a)
print(a)

#special functions 
print(np.sin(a))
print(np.cos(a))
print(np.log2(a))
```
